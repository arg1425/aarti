```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[1,2,7,5,8,2,4,1]
y2=[1,5,2,6,2,4,2,1]
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.set_title('plot line graph', fontsize=20)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.plot(x,y, label='curve1')
plt.plot(x,y2, label='curve2')
ax.legend()
```
